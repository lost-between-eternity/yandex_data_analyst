## Часть 1. Приоритизация гипотез.

### ICE

ICE (от англ. impact, confidence, effort / ease «влияние, уверенность, усилия / простота») — один из самых популярных способов приоритизации задач.

```python
hypothesis['ICE']=((hypothesis['impact']*hypothesis['confidence'])/hypothesis['efforts']).round(2)
hypothesis=hypothesis.sort_values(by='ICE')

#присвоим каждому значению очки в порядке веса критерия для построения графика сравнения:
hypothesis.index = range(len(hypothesis))
hypothesis_ICE = hypothesis[['hypothesis', 'ICE']].reset_index()
hypothesis_ICE.columns = ['ICE_score','hypothesis', 'ICE']
hypothesis_ICE['ICE_score'] += 1
hypothesis_ICE.sort_values(by='ICE')
```

### RICE

Модификация RICE учитывает охват (R от англ. reach, «охват»):

```python
hypothesis['RICE'] = ((hypothesis['reach']*hypothesis['impact']*hypothesis['confidence'])/hypothesis['efforts']).round(2)
hypothesis=hypothesis.sort_values(by='RICE')

#присвоим каждому значению очки в порядке веса критерия для построения графика сравнения:
hypothesis.index = range(len(hypothesis))
hypothesis_RICE = hypothesis[['hypothesis', 'RICE']].reset_index()
hypothesis_RICE.columns = ['RICE_score', 'hypothesis', 'RICE']
hypothesis_RICE['RICE_score'] += 1
hypothesis_RICE
```

### Сравнительный анализ использования ICE и RICE

__Выведем общую таблицу:__

```python
hypothesis.sort_values(by='RICE', ascending=False)
```

```python
#Визуализируем значимости:
comparison_of_hypotheses = hypothesis_RICE.merge(hypothesis_ICE, on='hypothesis').sort_values(by='RICE_score')
comparison_of_hypotheses = comparison_of_hypotheses[['hypothesis', 'RICE_score', 'ICE_score']]

comparison_of_hypotheses_pivot = comparison_of_hypotheses.pivot_table(
        columns = 'hypothesis',
        values= ('ICE_score','RICE_score'))
    
comparison_of_hypotheses.plot(kind = 'bar', 
                              figsize=(21,7), 
                              rot=0, 
                              grid=True, 
                              ylim=(0,10), 
                              colors=('crimson', 'darkblue'))
plt.legend(fontsize='xx-large')
plt.title('Сравнение RICE и ICE', fontsize='xx-large')
plt.ylabel('Вес значения', fontsize='large')
plt.xlabel('Гипотезы', fontsize='large')
plt.yticks(np.arange(0, 10.5, 1))
plt.show()

print('Значения:')
print(hypothesis['hypothesis'])
```

## Часть 2. Анализ A/B-теста.

### График кумулятивной выручки по группам.

```python
# Cоздаем массив уникальных пар значений дат и групп теста
datesGroups = orders[['date', 'group']].drop_duplicates()
# Агрегируем значения
# Получаем агрегированные кумулятивные по дням данные о заказах 
ordersAggregated = datesGroups.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transactionid': pd.Series.nunique,
            'visitorid': pd.Series.nunique,
            'revenue': 'sum'
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

# Получаем агрегированные кумулятивные по дням данные о посетителях
visitorsAggregated = datesGroups.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg({
            'date': 'max', 
            'group': 'max', 
            'visitors': 'sum'
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

# Объединяем кумулятивные данные в одной таблице и меняем названия столбцов
cumulativeData = ordersAggregated.merge(
    visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group']
)
cumulativeData.columns = [
    'date',
    'group',
    'orders',
    'buyers',
    'revenue',
    'visitors',
]

cumulativeRevenueA = cumulativeData[cumulativeData['group']=='A'][['date','revenue', 'orders']]
cumulativeRevenueB = cumulativeData[cumulativeData['group']=='B'][['date','revenue', 'orders']]
# Вывод
cumulativeData.head()
```

__Построим графики кумулятивной выручки по дням и группам:__ 

```python
plt.figure(figsize=(18,6))
plt.plot(cumulativeRevenueA['date'], cumulativeRevenueA['revenue'], 
         label='A', 
         color='darkblue', 
         marker = 'o', 
         linewidth=4, 
         alpha=0.75
        )
plt.plot(cumulativeRevenueB['date'], cumulativeRevenueB['revenue'], 
         label='B', 
         color='crimson', 
         marker = 'o', 
         linewidth=4, 
         alpha=0.75
        )

plt.legend(fontsize='xx-large') 
plt.grid()

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=1))
plt.gcf().autofmt_xdate(rotation=30)

plt.xlabel('Дата', fontsize='large')
plt.ylabel('Кумулятивная выручка', fontsize='large')
plt.title('График кумулятивной выручки', fontsize='xx-large')
plt.tight_layout()
plt.show()
```

Группа "В" показывает результат лучше, чем группа "А".  
18 августа наблюдается резкий рост значения выручки, вполне возможно это вызвано одной или несколькими большими покупками.

### Статистическая значимость различий в конверсии между группами по «сырым» данным.

__Нулевая гипотеза:__ статистически значимых различий в конверсии между группами A и B по "сырым данным" нет.  
__Альтернативная гипотеза:__ статистически значимые различия в конверсии между группами A и B по "сырым данным" есть.

__Применим тест Манна-Уитни__

```python
visitorsADaily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']
visitorsBDaily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']
ordersByUsersA = (
    orders[orders['group'] == 'A']
    .groupby('visitorid', as_index=False)
    .agg({'transactionid': pd.Series.nunique})
)
ordersByUsersA.columns = ['visitorid', 'orders']
 
ordersByUsersB = (
    orders[orders['group'] == 'B']
    .groupby('visitorid', as_index=False)
    .agg({'transactionid': pd.Series.nunique})
)
ordersByUsersB.columns = ['visitorid', 'orders']

sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsADaily['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsBDaily['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)


alpha = 0.05
p_val_conv = stats.mannwhitneyu(sampleA, sampleB)[1]
if (p_val_conv < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print('Cтатистическая значимость различий в конверсии между группами', end=" ") 
print('по "сырым" данным: {0:.5f}'.format(p_val_conv))
```

Принимаем альтернативную гипотезу: статистически значимые различия в конверсии между группами A и B по "сырым данным" есть.

__Нулевая гипотеза:__ статистически значимых различий в конверсии между группами A и B по "очищенным данным" нет.  
__Альтернативная гипотеза:__ статистически значимые различия в конверсии между группами A и B по "очищенным данным" есть.  

__Применим тест Манна-Уитни__

Сформируем датасет с учетом ранее выявленных аномалий.

```python
#cформируем перечень аномальных пользователей
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > 2]['visitorid'],
        ordersByUsersB[ordersByUsersB['orders'] > 2]['visitorid'],
    ],
    axis=0,
)
usersWithExpensiveOrders = orders[orders['revenue'] > 26785]['visitorid']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)

#сформируем "очищенные данные"
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['visitorid'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsADaily['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['visitorid'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                visitorsBDaily['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
alpha = 0.05
p_val_conv_filtered = stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1]
if (p_val_conv < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
    
print('Cтатистическая значимость различий в конверсии между группами', end=" ") 
print('по "очищенным" данным: {0:.5f}'.format(p_val_conv))
```

Принимаем альтернативную гипотезу: статистически значимые различия в конверсии между группами A и B по "очищенным данным" есть.
