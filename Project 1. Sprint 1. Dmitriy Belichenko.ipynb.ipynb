#### Пропуски трудового стажа

В процессе изучения таблицы в первом разделе были выявлены пустые значения и определено количество. Выявлено, что `dob_years` и `days_employed` пропущены одновременно. Поменяем на медианные значения с учетом возраста и типа занятости. 

Сформируем возврастные группы людей на основе 25%, 50%, 75% и отдельно выделим группу "пенсионеры" с возрастом более 60 лет.

```python
data['dob_years'].describe()
```

Создадим функцию, которая разделит на возрастные группы

```python
def age_group(age):
    if age <= 34:
        return 'молодой возраст'
    if age <= 43:
        return 'зрелый возраст'
    if age <= 53:
        return 'средний возраст'
    if age < 60:
        return 'пожилой возраст'
    else: 
        return 'пенсионеры'     
    
data['age_group'] = data['dob_years'].apply(age_group)
data.head(5)
```

Определим медианные значения по выделенным возрастным группам:

```python
median_days_employed = data.groupby(age_group)['days_employed'].median()
median_days_employed
```

Заменим пропуски на медианные значения по выделенным группам. 

```python
data.loc[(data['days_employed'].isnull()) & (data['age_group'] == 'молодой возраст'), 'days_employed'] = data.loc[data['age_group'] =='молодой возраст', 'days_employed'].median()
data.loc[(data['days_employed'].isnull()) & (data['age_group'] == 'зрелый возраст'), 'days_employed'] = data.loc[data['age_group'] =='зрелый возраст', 'days_employed'].median()
data.loc[(data['days_employed'].isnull()) & (data['age_group'] == 'средний возраст'), 'days_employed'] = data.loc[data['age_group'] =='средний возраст', 'days_employed'].median()
data.loc[(data['days_employed'].isnull()) & (data['age_group'] == 'пожилой возраст'), 'days_employed'] = data.loc[data['age_group'] =='пожилой возраст', 'days_employed'].median()
data.loc[(data['days_employed'].isnull()) & (data['age_group'] == 'пенсионер'), 'days_employed'] = data.loc[data['age_group'] =='пенсионер', 'days_employed'].median()
```

Проверяем исправления:

```python
data[(data['total_income'].isnull() == True) & (data['days_employed'].isnull() == True)].info()
```

Пропуски исправлены.

### Категоризация данных

#### Категоризация по доходу

Произведем категорицаю данных по показателю доходов людей. Hазделение произведем по квантилям, так мы получим группы почти одинакого размера. Возьмем разделение с шагом в 25%, которое мы уже видели в первом разделе исследования при изучении исходных данных.

```python
pd.qcut(data['total_income'], q=4).value_counts(sort=False)
```

Применим функцию для разделения на категории по выделенным квантилям:

```python
group_1 = 107623
group_2 = 142594
group_3 = 195820
group_4 = 2265604

def total_income_cut(total_income):
    if total_income <= group_1:
        return 'Группа 1'
    if total_income <= group_2:
        return 'Группа 2'
    if total_income <= group_3:
        return 'Группа 3'
    else:
        return 'Группа 4'
    
data['total_income_cut'] = data['total_income'].apply(total_income_cut)
data.sample(10)
```
